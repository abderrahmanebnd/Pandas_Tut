# Seaborn  
- Data Visualization Library : it give us meaning when we draw the plots for example

```python
# Importing the libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

Note: Seaborn has some built-in datasets

```python
# total bill vs tip dataset
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
# setting a theme for the plots
sns.set_theme()
```

```python
# visualize the data
sns.relplot(data=tips,x="total_bill",y='tip',col="time",hue="smoker",style="smoker",size="size")
```

```python
sns.get_dataset_names()
```

```python
# load the iris dataset
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

Scatter plot --> help us to cluster the data based on their parametrs

```python
sns.scatterplot(data=iris,x="sepal_length",y="petal_length",hue="species")
```

```python
# loading the titanic dataset
titanic = sns.load_dataset("titanic")
```

```python
titanic.head()
```

```python
titanic.shape
```

Count Plot

```python
sns.countplot(x="class",data=titanic)
```

```python
sns.countplot(x='survived',data=titanic)
```

Bar Chart

```python
sns.barplot(data=titanic,x='sex',y='survived',hue='class')
```

```python
# house price dataset
from sklearn.datasets import fetch_california_housing
housing_cl = fetch_california_housing()
type(housing_cl)
house  = pd.DataFrame(housing_cl.data,columns=housing_cl.feature_names)
house['PRICE'] = housing_cl.target
```

```python
house.head()
```

Distribution Plot

```python
sns.distplot(house.PRICE)
```

Correlation:
- Positive correlation
- Negative correlation

Heat Map

```python
correlation = house.corr()
```

```python
# Constructing a Heat Map
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt=".1f",annot=True,annot_kws={"size":10},cmap="Purples")
```
