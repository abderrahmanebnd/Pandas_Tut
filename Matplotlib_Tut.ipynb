Matplotlib:

- Useful for making Plots

```python
# importing matplotlib library -- Most of the Matplotlib utilities lies under the pyplot submodule
import matplotlib.pyplot as plt
```

```python
# import numpy library to get data for our plots
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

Pltting the data

```python
# sin wave
plt.plot(x,y)
plt.show()
```

```python
# cos wave
plt.plot(x,y)
plt.show()
```

```python
# adding title x-axis & y-axis labels
plt.plot(x,y)
plt.xlabel("Agle")
plt.ylabel("Sin value")
plt.title("Sin wave")
plt.show()
```

```python
# parabola
x = np.linspace(-10,10,20)
y = x**2;
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,"r+")
plt.show()
```

```python
plt.plot(x,y,"g.")
plt.show()
```

```python
plt.plot(x,y,"rx")
plt.show()
```

```python
x = np.linspace(-5,5,50)
plt.plot(x,np.sin(x),"g-")
plt.plot(x,np.cos(x),"r--")
plt.show()
```

Bar Plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1]) # creates an axes object (ax) within the figure (fig).
# The first two values [0, 0] specify the coordinates of the lower left corner of the axes in the figure.
# The last two values [1, 1] specify the width and height of the axes as a fraction of the figure's width and height.
langs =["en","fr","ar","lt"]
people = [100,50,60,30]
ax.bar(langs,people,width=0.5)
plt.xlabel("Languages")
plt.ylabel("Number of people")
plt.show()
```

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
langs =["en","fr","ar","lt"]
people = [100,50,60,30]
ax.pie(people,labels=langs,autopct="%.2f%%")
plt.legend()
plt.show()
```

Scatter Plot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color="b")
plt.show()
```

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap="Purples")
plt.show()
```
