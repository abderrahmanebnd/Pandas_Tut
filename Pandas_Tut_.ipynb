Pandas Library

Pandas Data Frame:
Pandas DataFrame is two-dimensional tabular data structure with labeled axes(rows and columns)

````python
# Importing the pandas and numpy libraries
import pandas as pd
import numpy as np
````

Creating a Pandas DataFrame

````python
# importing the california house price data
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
````

````python
print(housing)
````

````python
# pandas DataFrame
california_df = pd.DataFrame(housing.data,columns=housing.feature_names,)
````

````python
california_df.head(5)
````

````python
california_df.shape
````

Manipulating a DataFrame

````python
# adding a column to a dataframe
california_df["Price"] = housing.target
````

````python
california_df.head()
````

````python
# removing a row temporarly
california_df.drop(index=0,axis=0)
# axis = 0  if we want to remove a row , axis=1 if we want to remove a column
````

````python
# removing a column temporarly
california_df.drop(columns="MedInc",axis=1)
````

````python
# locating a row using the index value
california_df.iloc[0]
````

````python
# locating a particular column
print(california_df.iloc[:,1]) # first column
print(california_df.iloc[:,2]) # last column
````
